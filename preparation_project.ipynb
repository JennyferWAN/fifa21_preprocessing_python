```python
for colonne in df2.columns:
    doublons_colonne = df2[colonne].duplicated().any()

    if doublons_colonne:
        print(f"Colonne '{colonne}' a des doublons.")
```

```python
doublons_par_colonne = {}

for colonne in df2.columns:
    doublons_colonne = df2[colonne].duplicated().sum()
    doublons_par_colonne[colonne] = doublons_colonne

doublons_par_colonne_tries = dict(sorted(doublons_par_colonne.items(), key=lambda item: item[1], reverse=False))

for colonne, doublons in doublons_par_colonne_tries.items():
    print(f"Colonne '{colonne}' : {doublons} doublons.")
```
